```python
import numpy as np
```

```python
A = [[1,2,0], [2,0,2]]
```

```python
# do SVD for A
U, s, V = np.linalg.svd(A, full_matrices=True)
# U represents the left singular vectors
# s represents the singular values
# V represents the right singular vectors
print(U)
print(s)
print(V)
```

```python
# define a function to calculate svd of A
def svd(A):
    # calculate A^T * A
    ATA = np.dot(np.transpose(A), A)
    # calculate A * A^T
    AAT = np.dot(A, np.transpose(A))
    # calculate eigenvalues and eigenvectors of A^T * A
    eigenvalues_ATA, eigenvectors_ATA = np.linalg.eig(ATA)
    # calculate eigenvalues and eigenvectors of A * A^T
    eigenvalues_AAT, eigenvectors_AAT = np.linalg.eig(AAT)
    # sort eigenvalues and eigenvectors in descending order
    idx_ATA = eigenvalues_ATA.argsort()[::-1]
    eigenvalues_ATA = eigenvalues_ATA[idx_ATA]
    eigenvectors_ATA = eigenvectors_ATA[:,idx_ATA]
    idx_AAT = eigenvalues_AAT.argsort()[::-1]
    eigenvalues_AAT = eigenvalues_AAT[idx_AAT]
    eigenvectors_AAT = eigenvectors_AAT[:,idx_AAT]
    # calculate singular values
    s = np.sqrt(eigenvalues_ATA)
    # calculate left singular vectors
    U = eigenvectors_AAT
    # calculate right singular vectors
    V = eigenvectors_ATA
    # return U, s, V
    return U, s, V
```

```python
print(svd(A))
```
