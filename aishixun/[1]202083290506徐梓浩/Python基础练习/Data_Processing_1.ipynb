```python
fo = open("data.txt", "r")
# print(fo.read())
lines = fo.readlines()
for line in lines:
    print(line)
```

```python
# 把上述保存到字典中
dict = {}
for line in lines:
    k, v = line.split("，")
    dict[k] = v
print(dict)
```

```python
name = str(input("请输入姓名："))
print(dict.get(name, "查无此人"))
```
